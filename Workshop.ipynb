# Digital Image Processing Workshop

## Basics of Image Processing

### Reading an Image

```python
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('image_space.jpeg')

#Print image pixels
print(image)
```

### Displaying the image

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

### Extracting Height and Width of an image

```python
h , w = image.shape[:2]
print(f"Height : {h} , Width : {w}") 
```

### Extracting the RGB Values from image

OpenCV reads the pixels in the reverse order i.e BGR

```python
#Extracting RGB values at the position (101 , 0)

B , G , R = image[101 , 0]     #Here [101 , 0] is a position in the image i.e coordinates

print(f"R : {R} , G : {G} , B : {B}")
```

Extracting the R , G and B values seperately \
B = 0 , G = 1 and R = 2

```python
B = image[100 , 100 , 0]     #Blue component at position (100 , 100)
G = image[100 , 100 , 1]     #Green component at position (100 , 100)
R = image[100 , 100 , 2]     #Red component at position (100 , 100)

print(f"R : {R} , G : {G} , B : {B}")
```

### Region of Interest (ROI)

```python
roi = image[100:500 , 200:700]     #Row 100 to 500 and Column 200 to 700
plt.imshow(roi)
plt.axis("off")
plt.show()
```

```python
#ROI compoared to original image

plt.figure(figsize = (10 , 5))
plt.subplot(1 , 2 , 1)
plt.imshow(image)
plt.axis("off")
plt.title("Original image")

plt.subplot(1 , 2 , 2)
plt.imshow(roi)
plt.axis("off")
plt.title("ROI image")

plt.show()
```

### Resizing the image

```python
resized_image = cv2.resize(image , (500 , 500))
plt.axis("off")
plt.title("Resized image")
plt.imshow(resized_image)
plt.show()
```

### Drawing shapes in the image

```python
#First we will need to copy the original image

image_copy = image.copy()

#Drawing a rectangle in the image

rectangle = cv2.rectangle(image_copy , (418 , 388) , (1060 , 2000) , color = (0, 255, 0) , thickness = 10)
plt.axis("off")
plt.imshow(rectangle)
plt.show()
```

### Adding text in image

The function `cv2.putText` is used \
These are it's inputs 
1. Image
2. Text
3. Position
4. Font
5. Font Size
6. Font color
7. Fon Thickness
8. Line type

```python
image_copy2 = image.copy()
text = cv2.putText(image_copy2 , "Space is Awesome" , (1000 , 1000) , cv2.FONT_HERSHEY_COMPLEX 
                   , 6 , (0 , 255 , 0) , 6 , cv2.LINE_AA)

plt.imshow(text)
```
